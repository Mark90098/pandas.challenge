```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"]) #not sure why school name is there twice
school_data_complete.head()
```

```python
# Calculate the total number of unique schools
#school_count = school_data_complete[["school_name"]].value_counts() #returns the count of all the different data entries with each school
#school_count = (school_data_complete["school_name"]).unique() #returns each unique school name
school_count = (school_data_complete["school_name"]).nunique() #returns number of unique schools
school_count
```

```python
# Calculate the total budget
#total_budget = school_data_complete["budget"].sum() #budget data is repeated for every student, so I need to sum it once per school. I could simply call unmerged school_data or use groupby function
#total_budget = school_data_complete.loc[:,["school_name","budget"]] #seperates df columns but doesn't make unique budget data
groupby_school = school_data_complete.groupby(["school_name"])
groupby_school.mean() #returns mean of each group for each column. All budget data is the same for every group, so this makes all schools show up only once
```

```python
total_budget = groupby_school["budget"].mean() #turns the budget column into a series that I can add up
Total_budget = total_budget.sum()
Total_budget
```

```python
# Calculate the average (mean) math score

average_math_score = school_data_complete["math_score"] #turns math score column into series (from orginal df, this is overall average accross all data)
average_math_score = average_math_score.mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"] #turns math score column into series (from orginal df, this is overall average accross all data)
average_reading_score = average_reading_score.mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"] #this last bracket makes the calculation only happen for 1 column. Otherwise, all columns would return 74.9
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"] #this last bracket makes the calculation only happen for 1 column. Otherwise, all columns would return 74.9
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total Schools":[school_count], "Total Students": [student_count],"Total Budget": [Total_budget],"Average Math Score": [average_math_score],"Average Reading Score": [average_reading_score],"% Passing Math": [passing_math_percentage],"% Passing Reading": [passing_reading_percentage], "% Passing Overall": [overall_passing_rate]})
district_summary

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```
